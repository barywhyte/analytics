**Twitter often returns duplicate results, we can filter out duplicate results**

```python
all_text = []
filtered_statuses = []
for s in statuses:
    if s['text'] not in all_text:
        filtered_statuses.append(s)
        all_text.append(s['text'])
statuses_f = filtered_statuses
```

**6. Extracting text, screen names, and hashtags from tweets**

```python
status_text = [status['text'] for status in statuses_f]

screen_name = [user_mention['screen_name']
                  for status in statuses_f
                     for user_mention in status['entities']['user_mentions']
              ]

hashtag = [hashtag['text']
             for status in statuses_f
                for hashtag in status['entities']['hashtags']
          ]
#compute a collection of all words from all tweets

words = [w for t in status_text
        for w in t.split()     
    ]
```

**7. Creating basic frequency distribution from words in tweets.**

```python
from collections import Counter
```

```python
for item in [words, screen_name, status_text, hashtag]:
    c = Counter(item)
    print(c.most_common()[:10])
    print()
```

**Create a pettyprint function to display turples in a nice tabular format**

```python
def prettyprint(label, list_of_tuple):
    print('\n{:^20} | {:^6}'.format(label, 'count'))
    print('*'*40)
    for k,v in list_of_tuple:
        print('{:20} | {:6}'.format(k,v))
```

```python
for label, data in (('Words', words),
                    ('ScreenName', screen_name),
                    ('Hashtag', hashtag)):
    c = Counter(data)
    prettyprint(label, c.most_common()[:10])
```

**Build another prettyprint function to print entire tweents with their retweet counts**

```python
row_template = "{:7} | {:15} | {:50}"
def prettyprint_tweets(list_of_tuples):
    print()
    print(row_template.format('Count', 'Screeb Name', 'Text'))
    print('*'*80)
    for count, screen_name, text in list_of_tuples:
        print(row_template.format(count, screen_name, text[:50]))
        if len(text) > 50:
            print(row_template.format("","", text[50:100]))
            if len(text) > 100:
                print(row_template.format("", "", text[100:]))
```

```python
prettyprint_tweets(sorted(retweets, reverse=True)[:10])
```
